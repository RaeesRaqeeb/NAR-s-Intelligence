```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Sample student data with multiple attributes
data = {
    'Physics': [85, 45, 60, 92, 40, 78, 55, 88, 50, 72],
    'Math': [90, 50, 55, 88, 42, 81, 60, 85, 48, 70],
    'English': [88, 48, 58, 85, 45, 80, 53, 84, 50, 65],
    'Attempts': [1, 3, 2, 1, 4, 2, 3, 1, 4, 2],
    'Time': [30, 60, 45, 25, 70, 40, 50, 35, 65, 55]
}

# Convert data to a DataFrame
df = pd.DataFrame(data)
print("Student Data:\n", df)

# Prepare data for clustering (Physics, Math, English, Attempts, Time)
X = df.values

# Apply K-means clustering with 3 clusters (High, Average, Low performers)
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(X)

# Display the clusters
print("\nClustered Data:\n", df)

# Visualizing the clustering using Physics and Math scores (2D plot)
plt.figure(figsize=(8, 6))
plt.scatter(df['Physics'], df['Math'], c=df['Cluster'], cmap='viridis', s=100, marker='o')
plt.title('Student Clustering based on Physics and Math Scores')
plt.xlabel('Physics Marks')
plt.ylabel('Math Marks')
plt.colorbar(label='Cluster')
plt.show()
```
