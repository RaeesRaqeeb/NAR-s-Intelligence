University prediction models DataSet


```python
import pandas as pd
import numpy as np
import random

np.random.seed(42)  # For reproducibility

num_samples = 10000

physics_marks = np.random.randint(30, 100, num_samples)
math_marks = np.random.randint(20, 100, num_samples)
english_marks = np.random.randint(40, 100, num_samples)

universities = ['University A', 'University B', 'University C', 'University D', 'University E']
university = [random.choice(universities) for _ in range(num_samples)]

admission_status = np.random.choice(['Admitted', 'Rejected'], num_samples, p=[0.6, 0.4]) #60% chance of admission
total_attempts = np.random.randint(1, 4, num_samples)

data_university = pd.DataFrame({
    'physics_marks': physics_marks,
    'math_marks': math_marks,
    'english_marks': english_marks,
    'university': university,
    'admission_status': admission_status,
    'total_attempts': total_attempts
})

data_university.to_csv('university_recommendation_data.csv', index=False)
print("University Recommendation Dataset Created.")
(data_university)
```

Weak Area Identification

```python
import pandas as pd
import numpy as np
import random

np.random.seed(42)

num_samples = 10000

physics_marks = np.random.randint(30, 100, num_samples)
math_marks = np.random.randint(30, 100, num_samples)
english_marks = np.random.randint(30, 100, num_samples)

physics_difficulty = np.random.randint(1, 6, num_samples) #difficulty scale 1 to 5

physics_time = np.random.uniform(10, 60, num_samples)  # Time in seconds
math_time = np.random.uniform(10, 60, num_samples)
english_time = np.random.uniform(10, 60, num_samples)

physics_correct = np.random.choice([True, False], num_samples, p=[0.7, 0.3]) #70% correct
math_correct = np.random.choice([True, False], num_samples, p=[0.6, 0.4])
english_correct = np.random.choice([True, False], num_samples, p=[0.8, 0.2])

data_weak_area = pd.DataFrame({
    'physics_marks': physics_marks,
    'math_marks': math_marks,
    'english_marks': english_marks,
    'physics_difficulty' : physics_difficulty,
    'physics_time': physics_time,
    'math_time': math_time,
    'english_time': english_time,
    'physics_correct': physics_correct,
    'math_correct': math_correct,
    'english_correct': english_correct
})

data_weak_area.to_csv('weak_area_data.csv', index=False)
print("\nWeak Area Identification Dataset Created.")
data_weak_area
```

Student categorization

```python
import pandas as pd
import numpy as np

np.random.seed(42)

num_samples = 10000

physics_marks = np.random.randint(30, 100, num_samples)
math_marks = np.random.randint(30, 100, num_samples)
english_marks = np.random.randint(30, 100, num_samples)

total_score = physics_marks + math_marks + english_marks

data_student_category = pd.DataFrame({
    'physics_marks': physics_marks,
    'math_marks': math_marks,
    'english_marks': english_marks,
    'total_score': total_score
})

data_student_category.to_csv('student_category_data.csv', index=False)
print("\nStudent Categorization Dataset Created.")
(data_student_category)
```

The above dataset is too uniform so we use different method to create realistic dataset

```python
import numpy as np
import pandas as pd
num_samples = 10000

# Simulating real-world variation in test scores
math_marks = np.random.normal(loc=60, scale=20, size=num_samples).clip(20, 99)  # Most scores near 60
physics_marks = np.random.normal(loc=55, scale=22, size=num_samples).clip(20, 99)  # Slightly harder subject
english_marks = np.random.normal(loc=70, scale=18, size=num_samples).clip(40, 99)  # Generally higher scores

df={'Math_marks':math_marks,'Physics_marks':physics_marks,'english_marks':english_marks}
df=pd.DataFrame(df)
```

```python
df.describe()
```

```python
# Create a correlation factor
math_marks = np.random.normal(loc=60, scale=20, size=num_samples)
physics_marks = math_marks * 0.8 + np.random.normal(loc=10, scale=10, size=num_samples)  # Physics correlates with Math
english_marks = 100 - math_marks * 0.3 + np.random.normal(loc=5, scale=10, size=num_samples)  # English is inversely related
df={'Math_marks':math_marks,'Physics_marks':physics_marks,'english_marks':english_marks}
df=pd.DataFrame(df)
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)  # For reproducibility

# Generate math marks (normally distributed)
math_marks = np.random.normal(loc=65, scale=10, size=1000)

# Create correlated physics marks with some randomness
physics_marks = math_marks + np.random.normal(loc=0, scale=5, size=1000)

# Clip values to stay within valid range
math_marks = np.clip(math_marks, 30, 100)
physics_marks = np.clip(physics_marks, 30, 100)

# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(physics_marks, math_marks, alpha=0.5)  # Add transparency for density
plt.xlabel("Physics Marks")
plt.ylabel("Math Marks")
plt.title("Scatter Plot of Math vs. Physics Marks")
plt.show()
```

```python
correlation_factor = 0.8  # Higher means stronger correlation
physics_marks = correlation_factor * math_marks + (1 - correlation_factor) * np.random.normal(65, 10, 1000)
```
